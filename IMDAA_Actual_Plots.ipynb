# Mean Sea Level Pressure (Code edited for better looking)

```python
for j in range(16,22):
    data=xr.open_dataset("C://Users/sreega/Desktop/Workstation/EPE_Identification/decEPE/IMDAA/ncum_imdaa_reanl_HR_PRMSL-msl_2023121500-2023122123.nc")
    # Choose the time of interest
    time = f"2023-12-{j}"
    min_lon, max_lon = 65, 95   # Example longitude range
    min_lat, max_lat = 5, 28    # Example latitude range
    # Convert the time to datetime format
    data['time'] = pd.to_datetime(data['time'].values)
    # Select the data for the specific date
    date_data = data.sel(time=time,latitude=slice(min_lat,max_lat),longitude=slice(min_lon,max_lon))
    date_data=date_data['PRMSL_msl'].resample(time="d").mean()

    anomaly_precp=date_data/100

    for i in range(len(anomaly_precp)):
        anomaly_precp1=anomaly_precp[i]
        
        label=np.datetime_as_string(anomaly_precp1['time'].values,unit='s').split('T')
        # Plot the results

        fig=plt.figure(figsize=(7,6))

        # Set the projection to PlateCarree for a global map
        ax = plt.axes(projection=ccrs.PlateCarree())

        # Add features to the map
        ax.coastlines(resolution='10m', color='black', linewidth=1)
        ax.add_feature(cfeature.LAND, facecolor='#F5F5DC')
        ax.add_feature(cfeature.OCEAN, facecolor='#E0FFFF')


        # Set plot extent (min/max lat/lon)
        min_lon, max_lon = 65, 95   # Example longitude range
        min_lat, max_lat = 5, 28    # Example latitude range
        ax.set_extent([min_lon, max_lon, min_lat, max_lat], crs=ccrs.PlateCarree())

        # Set aspect ratio
        ax.set_aspect('auto')  # Change to 'equal' if needed
        # Normalize the color scale
        norm = Normalize(vmin=anomaly_precp1.min().values, vmax=anomaly_precp1.max().values)
        levels=np.linspace(anomaly_precp1.min().values, anomaly_precp1.max().values,8).astype(int)
        contour = ax.contourf(anomaly_precp1['longitude'], anomaly_precp1['latitude'], 
                                anomaly_precp1.values, cmap='coolwarm',
                                levels=levels, transform=ccrs.PlateCarree(), extend='both')

        ## Add contour lines
        ax.contour(anomaly_precp1['longitude'], anomaly_precp1['latitude'], 
                anomaly_precp1.values, colors='black', linewidths=0.3, levels=levels,
                transform=ccrs.PlateCarree(), extend='both')


        # Create and customize the colorbar
        cbar = plt.colorbar(contour, ax=ax, pad=0.05, orientation='horizontal', extend='both')
        cbar.set_label('Mean Sea Level Pressure $hPa$')

        # Define coordinates and size for the square box
        lon_min, lon_max = 75, 80  # Example longitude bounds for the box
        lat_min, lat_max = 8, 12  # Example latitude bounds for the box

        # Create the box (Rectangle)
        box = Rectangle(
            (lon_min, lat_min),  # Lower-left corner
            lon_max - lon_min,   # Width
            lat_max - lat_min,   # Height
            edgecolor='red', facecolor='none', linewidth=2, transform=ccrs.PlateCarree()
        )

        # Add the box to the plot
        ax.add_patch(box)

        # Update title
        plt.title(f'Mean Sea Level Pressure on Date: {label[0]} {label[1]}')

        # Add gridlines and draw labels
        gl = ax.gridlines(draw_labels=True, linewidth=0.5, color='gray', alpha=0.5, linestyle='--')
        gl.top_labels = False
        gl.right_labels = False
        plt.tight_layout()
        #plt.savefig(f"C://Users/sreega/Desktop/Workstation/EPE_Identification/decEPE/IMDAA/MSLP/mslp_{label[0]}_{label[1][:2]}.png",dpi=300)
        #plt.close(fig)
```

# Temperature

```python
# Define the height level
height = 500

for j in range(16, 22):
    # Load the dataset
    data = xr.open_dataset("C://Users/sreega/Desktop/Workstation/EPE_Identification/decEPE/IMDAA/ncum_imdaa_reanl_HR_TMP-prl_2023121500-2023122121.nc")
    
    # Choose the time of interest
    time = f"2023-12-{j}"

    # Convert the time to datetime format
    data['time'] = pd.to_datetime(data['time'].values)

    # Select the data for the specific date and pressure level
    date_data = data.sel(time=time, plevel=height)

    # Resample to 6-hourly average
    date_data = date_data['TMP_prl'].resample(time="6h").mean()
    anomaly_precp = date_data

    for i in range(len(anomaly_precp)):
        anomaly_precp1 = anomaly_precp[i]
        label = np.datetime_as_string(anomaly_precp1['time'].values, unit='s').split('T')

        # Plot the results
        fig = plt.figure(figsize=(12, 5))

        # Set the projection to PlateCarree for a global map
        ax = plt.axes(projection=ccrs.PlateCarree())

        # Add map features
        ax.coastlines(resolution='10m', color='black', linewidth=1)
        ax.add_feature(cfeature.LAND, facecolor='#F5F5DC')
        ax.add_feature(cfeature.OCEAN, facecolor='#E0FFFF')

        # Set plot extent (min/max lat/lon)
        min_lon, max_lon = 40, 120  # Example longitude range
        min_lat, max_lat = -5, 40   # Example latitude range
        ax.set_extent([min_lon, max_lon, min_lat, max_lat], crs=ccrs.PlateCarree())

        # Normalize the color scale
        norm = Normalize(vmin=anomaly_precp1.min().values, vmax=anomaly_precp1.max().values)

        # Plot filled contours
        contour = ax.contourf(anomaly_precp1['longitude'], anomaly_precp1['latitude'], 
                              anomaly_precp1.values, cmap='coolwarm', norm=norm, transform=ccrs.PlateCarree())

        # Add contour lines
        ax.contour(anomaly_precp1['longitude'], anomaly_precp1['latitude'], 
                   anomaly_precp1.values, colors='black', linewidths=0.3, transform=ccrs.PlateCarree())

        # Create and customize the colorbar
        cbar = plt.colorbar(contour, ax=ax, orientation='vertical')
        cbar.set_label('Temperature (K)')

        # Define coordinates and size for the square box
        lon_min, lon_max = 75, 80  # Example longitude bounds for the box
        lat_min, lat_max = 8, 12   # Example latitude bounds for the box

        # Create and add the box (Rectangle)
        box = Rectangle((lon_min, lat_min), lon_max - lon_min, lat_max - lat_min, 
                        edgecolor='red', facecolor='none', linewidth=2, transform=ccrs.PlateCarree())
        ax.add_patch(box)

        # Update title
        plt.title(f'Temperature at Z{height} on Date: {label[0]} {label[1]}')

        # Add gridlines with labels
        gl = ax.gridlines(draw_labels=True, linewidth=0.5, color='gray', alpha=0.5, linestyle='--')
        gl.top_labels = False
        gl.right_labels = False

        plt.savefig(f"C://Users/sreega/Desktop/Workstation/EPE_Identification/decEPE/IMDAA/temperature/{height}/tmp_{label[0]}_{label[1][:2]}.png",dpi=300)
        plt.close(fig)
```
